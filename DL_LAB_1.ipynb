```python
# MCP Neuron OR Gate Implementation
import numpy as np
def func1(x,w, theta):
  return 1 if np.dot(x,w)>=theta else 0

w=np.array([1,1])
theta=1
for i in [(0,0),(0,1),(1,0),(1,1)]:
  print(func1(i,w,theta))
```

```python
# Perceptron implementation for AND gate (generalized for n inputs)

def perceptron(inputs, weights, bias):
    weighted_sum = 0
    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]
    weighted_sum += bias
    if weighted_sum >= 0:
        return 1
    else:
        return 0
n = int(input("Enter number of inputs: "))
inputs = list(map(int, input("Enter inputs (0 or 1): ").split()))
weights = [1] * n
bias = -(n - 0.5)
output = perceptron(inputs, weights, bias)
print("Output:", output)
```

```python
# Perceptron implementation for OR gate (generalized for n inputs)

def perceptron(inputs, weights, bias):
    weighted_sum = 0
    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]
    weighted_sum += bias
    if weighted_sum >= 0:
        return 1
    else:
        return 0
n = int(input("Enter number of inputs: "))
inputs = list(map(int, input("Enter inputs (0 or 1): ").split()))
weights = [1] * n
bias = -0.5
output = perceptron(inputs, weights, bias)
print("Output:", output)
```
